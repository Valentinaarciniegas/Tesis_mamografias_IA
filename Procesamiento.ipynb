```python
import os
import sys
from skimage import io
import numpy as np
import cv2
from skimage.morphology import dilation, erosion
from skimage.morphology import disk
import matplotlib.pyplot as plt
```

```python
def process_images():
    main_directory = r"C:\Users\hp\Documents\CursoML\allmias"
    CATEGORY = ["Training", "Testing"]
    
    output = []
    
    for category in CATEGORY: 
        path = os.path.join(main_directory, category)
        print("\n" + path)
        
        for folder in os.listdir(path):
            print("\nFolder: " + folder)
            
            for file in os.listdir(os.path.join(path, folder)):
                img_path = os.path.join(os.path.join(path, folder), file)
                print("\nProcessing: " + img_path)
                image = io.imread(img_path, as_gray=True, plugin='matplotlib')
                kernel = np.ones((3,3), np.uint8)
                eroded = cv2.erode(image, kernel, iterations=1)
                dilated = cv2.dilate(eroded, kernel, iterations=1)
                new_img_path = img_path.replace("allmias", "processed")
                multiplier = 255
                if "jpg" in new_img_path: multiplier = 1
                cv2.imwrite(new_img_path, dilated*multiplier)
                print("Saved in: " + new_img_path)
                
    print("\nFinished...")

process_images()
```

```python
def plot_comparison(original, filtered, original_name, filter_name):
    
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title(original_name)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(filter_name)
    ax2.axis('off')
```
